### 📥 Loading Dataset
The dataset is loaded using `read_csv()` and stored in a DataFrame for further analysis.

```python
df = pd.read_csv("Bank_churn.csv")
df
```

### 🔍 Previewing the dataset
- head() shows first 5 rows
- tail() shows last 5 rows

```python
df.head() # the number shows how many rows from above to be displayed, by default 5.
```

```python
df.tail() # the number shows how many bottom rows to be displayed, by default 5.
```

### 📊 Understanding dataset structure
- shape → rows & columns
- columns lists all feature names  
- info() → data types and missing values

```python
df.shape
```

```python
df.info()
```

### 📈 Statistical Summary

- Includes metrics like mean, min, max, and standard deviation  
- Useful for understanding data distribution  

```python
df.describe()
```

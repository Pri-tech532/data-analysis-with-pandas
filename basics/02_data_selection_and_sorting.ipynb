# 📊 Data Selection & Sorting with Pandas

## 📌 Objective
The objective of this notebook is to explore different techniques for selecting and accessing data using Pandas, including column selection, row selection, and conditional filtering.

```python
import pandas as pd
df = pd.read_csv("Bank_churn.csv")
```

### 📊 Data Types Overview
- Displays the data type of each column  
- Useful for understanding which columns are numerical or categorical before selection  

```python
df.dtypes
```

### 📊 Column Selection
- Selecting a single column returns a Series  
- Selecting multiple columns returns a DataFrame  
- Helps isolate relevant features for analysis  

```python
# Single column: to obtain series we use single []
df["Surname"]
```

```python
# Multiple columns: to obtain series we use double[[]]
df[["Surname", "CreditScore"]]
```

```python
df[["Surname", "CreditScore", "Geography"]]
```

### 🔄 Data Sorting

- Sorting is used to organize data based on one or more columns  
- Helps identify top or bottom records and understand data distribution  

**Types of Sorting:**
- Sorting by a single column  
- Sorting by multiple columns  
- Custom sorting order using ascending/descending values  

**Key Points:**
- By default, sorting is in ascending order  
- `ascending=False` is used for descending order  
- Multiple columns can be sorted with different sort orders  

Sorting improves data readability and prepares it for deeper analysis.

```python
# Sort by single column
df.sort_values("Age")
```

```python
# Sort multiple columns
df.sort_values(["Age", "CreditScore"])
```

```python
# Custom sorting order
df.sort_values(["Age", "CreditScore", "Tenure"],ascending=[False, True, True])
```

## 💡 Observations

- Column selection helps isolate relevant features for focused analysis  
- The dataset contains both categorical and numerical variables  
- Sorting allows identification of top or bottom records based on specific criteria  
- Multi-column sorting enables deeper insights by applying hierarchical ordering  
- Combining selection and sorting improves data readability and analysis efficiency

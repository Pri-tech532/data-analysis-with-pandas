# 🧹 Data Cleaning with Pandas

## 📌 Objective
The objective of this notebook is to clean the dataset by identifying and handling duplicate records, missing values, and basic data inconsistencies.

```python
import pandas as pd
df = pd.read_csv("Bank_churn.csv")
```

### 🔍 Duplicate Records
- `duplicated()` identifies repeated rows  
- `sum()` gives total number of duplicate records  

```python
# Check duplicate rows
df.duplicated()
```

```python
# Count duplicates
df.duplicated().sum()
```

### 🔎 Duplicate Analysis on Selected Columns

Instead of checking duplicates across the entire dataset, we can focus on specific columns to identify repeated patterns more effectively.

```python
# Selecting relevant columns for duplicate analysis
subset_df = df[["Surname", "CreditScore", "Geography"]]
```

```python
# Extracting duplicate rows based on selected columns
subset_duplicates = subset_df[subset_df.duplicated()]
subset_duplicates
```

```python
#print("Remaining duplicates:", subset_df.duplicated().sum()) we can use print statement also to print the sum
subset_df.duplicated().sum()
```

```python
# Removing duplicate rows from the subset
subset_df.drop_duplicates(inplace=True)
```

```python
# Checking if duplicates still exist
subset_df.duplicated().sum()
```

### 🧹 Removing Duplicates
- Entire row duplicates can be removed  
- Specific columns can also be targeted using `subset`  

```python
# Creating a copy of original dataset
df_copy = df.copy()
```

```python
# Removing duplicates based on 'Balance' column
df_copy.drop_duplicates(subset="Balance", inplace=True)
df_copy
```

### ⚠️ Missing Values
- `isna()` helps identify missing data  
- `fillna()` is used to handle missing values  

```python
z = df.describe(include="all")
```

```python
z # here we have a lot of NaN values
```

```python
# Fill missing values with 0
z = z.fillna(0)
```

### 🔄 Data Type Conversion
- Converting data types ensures consistency  
- Useful when identifiers should not be treated as numerical values  

```python
df.info()
```

```python
df['CustomerId'] = df['CustomerId'].astype("str")
df.info()
```
